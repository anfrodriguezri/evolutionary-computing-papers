```python
import numpy as np

import evolutionary_computing.evolutionary_strategies as es
import evolutionary_computing.representation as rp
import utils.functions as fs

import matplotlib.pyplot as plt
import utils.ploting as ploting
%matplotlib inline


def run_experiment(optimization_function, selection_type):
    epochs = 10
    max_iter = 100

    fitness = np.zeros((epochs, max_iter))
    performance = np.zeros(epochs)
    
    if selection_type == 'Comma':
        selection = ','
    elif selection_type == 'Plus':
        selection = '+'

    for e in range(epochs):
        evo = es.EvolutionaryStrategy(optimization_function=optimization_function,
                                    population_size=100,
                                     total_parents=2,
                                     selection_type=selection,
                                     total_offspring=120,
                                     max_iterations=max_iter)
        fitness[e] = evo.run()

    #print(fitness)
    dimentions = optimization_function.dim
    title = f'ES: {optimization_function.name}' \
            + ' ($\mathbb{R}^{' + str(dimentions) + '}$) ' \
            + f'{selection_type} selection'
    
    ploting.plot(fitness, 
                 title=title, 
                 name=f'{optimization_function.__class__.__name__}_{dimentions}_{selection_type}')
```

```python
rastrigin = fs.RastriginFunction(2)

run_experiment(rastrigin, 'Plus')
run_experiment(rastrigin, 'Comma')

rastrigin = fs.RastriginFunction(3)

run_experiment(rastrigin, 'Plus')
run_experiment(rastrigin, 'Comma')
```
